```python
null = titanic["fare"].isnull().sum()
display(null)
median = copy["fare"].median()
copy["Med"] = copy['fare'].fillna(median).round(0)
display(copy)
```

```python
copy["pclass"] = copy['pclass'].replace({1:"1st",2:"2nd",3:"3rd"})
copy.sort_values("pclass",ascending = False).reset_index(drop=True)
```

```python
copy["survived"] = copy["survived"].replace({0:"No",1:"Yes"})
```

```python
copy["name"] = copy['name'].str.split(",").str[0]
copy["SURName"] = copy['name'].str.split(",").str[1]
```

```python
copy["embarked"] = copy['embarked'].replace({"C":"Cherbourg","Q":"Queenstown","S":"Southampton"})
copy.sort_values("embarked", ascending= True).reset_index(drop=True)
```
